Os modelos pré-treinados em geral possuem uma codificação simples para realizar as tarefas. Mas embora estejamos trabalhando com modelos de IA Generativa, podemos modificar algumas coisas no processo.


###Visão Geral
* Tarefa deste notebook: `text-generation`
* Modelo utilizado: `pierreguillou/gpt2-small-portuguese`
* GPT-2 é um modelo de linguagem generativa desenvolvido pela OpenAI, baseado na arquitetura Transformer, capaz de gerar texto coerente e relevante a partir de uma entrada inicial.


```python
#importando bibliotecas
import transformers
from transformers import pipeline
```

```python
gerador = pipeline("text-generation", model="pierreguillou/gpt2-small-portuguese")
texto = "A inteligência artificial está transformando o mundo de várias maneiras."
```

```python
#passando parâmetros
resultado = gerador(
    text_inputs = texto,
    max_length=100,
    do_sample=True,
    temperature=0.7,
    top_k=50,
    top_p=0.85,
    num_return_sequences=3
)
```

```python
for i, sequencia in enumerate(resultado):
    print(f"Sequência {i+1}:\n{sequencia['generated_text']}\n")
```
