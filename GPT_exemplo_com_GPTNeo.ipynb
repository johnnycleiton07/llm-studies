#GPT na prática com GPTNeo

GPT é uma série de modelos de linguagem desenvolvidos pela OpenAI, conhecidos por sua capacidade de gerar texto coerente e relevante, completar frases, responder perguntas, e muito mais.

* Os modelos GPT são baseados na arquitetura Transformer, mas se destacam por seu foco na geração de texto e não apenas na compreensão.

O objetivo desse notebook é completar texto.

##Configurações iniciais

```python
from transformers import GPTNeoForCausalLM, GPT2Tokenizer
```

* O modelo a ser utilizado será o GPTNeo de 125M da empresa EleutherAI, é um modelo livre para uso.

```python
model_name = 'EleutherAI/gpt-neo-125M'
model = GPTNeoForCausalLM.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
```

```python
#tokenização
prompt = "O estudo de Inteligência Artificial é"
inputs = tokenizer.encode(prompt, return_tensors='pt')
print(inputs)
```

##Gerando a saída

```python
#codigo de geração da saída
outputs = model.generate(
        inputs,
        max_new_tokens=20,
        temperature=0.7,
        top_k=50,
        top_p=0.95,
        num_return_sequences=1,
        pad_token_id=tokenizer.eos_token_id,
        no_repeat_ngram_size=2  ,
        do_sample=True
    )
```

```python
print(outputs)
```

```python
#decodificando os tokens para a saída em texto
text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(text)
```

* Para melhorar as respostas do modelo, é necessário ir modificando o código de geração de saída, modificando a quantidade de tokens, etc.
