###Visão Geral
* Modelo utilizado: `Helsinki-NLP/opus-mt-tc-big-en-pt`
* OPUS: Focado em tradução automática entre múltiplos idiomas, ele é um **recurso de corpus paralelo** e utiliza modelos de tradução neural baseados na arquitetura Transformer.

```python
#instalando ferramenta de tokenização usada em NLP
!pip install sacremoses
```

```python
#importando bibliotecas
from transformers import MarianMTModel, MarianTokenizer
```

```python
model_name = "Helsinki-NLP/opus-mt-tc-big-en-pt"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
texto_ingles = "Artificial intelligence is transforming the world in many ways."
src_text = f">>por<< {texto_ingles}"
tokens = tokenizer(src_text, return_tensors="pt", padding=True)
```

```python
traducao = model.generate(**tokens, max_length=100, num_return_sequences=3)
traducoes = [tokenizer.decode(t, skip_special_tokens=True) for t in traducao]
```

```python
#tradução vetorizada e tokenizada
print(traducao)
```

```python
#tradução em texto
for i, t in enumerate(traducoes):
    print(f"Tradução {i+1}: {t}")
```
