Os modelos pré-treinados em geral possuem uma codificação simples para realizar as tarefas. Mas embora estejamos trabalhando com modelos de IA Generativa, podemos modificar algumas coisas no processo.


###Visão Geral
* Tarefa deste notebook: `question-answering`
* Modelo utilizado: `deepset/roberta-base-squad2`
* RoBERTa (Robustly Optimized BERT Approach) é uma versão otimizada do modelo BERT, desenvolvida pelo Facebook AI, que melhora o desempenho em tarefas de compreensão de linguagem natural através de ajustes no pré-treinamento e maior quantidade de dados.

```python
#importando bibliotecas
import transformers
from transformers import pipeline
```

```python
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")
```

```python
contexto = """
A inteligência artificial (IA) refere-se à simulação da inteligência humana
em máquinas que são programadas para pensar como humanos e imitar suas ações.
O termo também pode ser aplicado a qualquer máquina que exiba traços associados
a uma mente humana, como aprendizagem e resolução de problemas.
"""
pergunta = "Quais são os traços associados a uma mente humana que a IA pode exibir?"
```

```python
#passando parâmetros
respostas = qa_pipeline(
    question=pergunta,
    context=contexto,
    max_answer_length=50,
    handle_impossible_answer=True,
    top_k=3
)
```

```python
for i, resposta in enumerate(respostas):
    print(f"Resposta {i+1}: {resposta['answer']} (Score: {resposta['score']})")
```
