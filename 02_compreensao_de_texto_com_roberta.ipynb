Os modelos pré-treinados em geral possuem uma codificação simples para realizar as tarefas. Mas embora estejamos trabalhando com modelos de IA Generativa, podemos modificar algumas coisas no processo.


###Visão Geral
* Tarefa deste notebook: `question-answering`
* Modelo utilizado: `deepset/roberta-base-squad2`
* RoBERTa (Robustly Optimized BERT Approach) é uma versão otimizada do modelo BERT, desenvolvida pelo Facebook AI, que melhora o desempenho em tarefas de compreensão de linguagem natural através de ajustes no pré-treinamento e maior quantidade de dados.

```python
#importando bibliotecas
import transformers
from transformers import pipeline
```

```python
qa = pipeline("question-answering", model="deepset/roberta-base-squad2")
```

```python
contexto = """
Carl Edward Sagan foi um cientista, astrônomo, astrobiólogo, astrofísico,
escritor, divulgador científico e ativista norte-americano.
Sagan é autor de mais de 600 publicações científicas e também de mais de
vinte livros de ciência e ficção científica.
"""
pergunta = "Quem foi Carl Sagan"
```

```python
#passando parâmetros
resposta = qa(
    question=pergunta,
    context=contexto,
    max_answer_length=50,
    handle_impossible_answer=True,
    top_k=3
)
```

```python
for i, r in enumerate(resposta):
    print(f"Resposta {i+1}: {r['answer']} (Score: {r['score']})")
```
