#Geração de texto com modelo de LLM da Hugging Face

Este projeto conta com a aplicação de um modelo de LLM da empresa 🤗 **Hugging Face** baseado na arquitetura **Transformers**. Basicamente, ele recebe um tema e desenvolve um texto alinhado a ele.

* Através da biblioteca `transformers`, o modulo `pipeline` será responsável por executar uma tarefa de PLN com uma alta abstração.


```python
!pip install transformers
```

```python
import transformers
from transformers import pipeline
```

É necessário ir a biblioteca da [Hugging Face](https://huggingface.co/) para encontrar um modelo que seja compatível com a tarefa que deseja executar. Nesse projeto, será usada a biblioteca `pierreguillou/gpt2-small-portuguese`. Para ter acesso a ela é só entrar no site e seguir os passos:

* Hugging Face > Models > em Tasks vai em Natural Language Processing > seleciona Text Generation > em Languages seleciona Portuguese.

Caso não coloque um modelo, será utilizado um modelo padrão que também funciona para a tarefa, porém a precisão (score) muda.

```python
gerador = pipeline("text-generation", model="pierreguillou/gpt2-small-portuguese")
```

```python
texto = "O gênero musical Rock atualmente"
resultado = gerador(texto, max_length=200, do_sample=True)
print(resultado)
```
