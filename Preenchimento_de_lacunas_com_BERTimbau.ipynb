#Preenchimento de lacunas com BERTimbau

Este projeto conta com a aplicação de um modelo de LLM da empresa 🤗 **Hugging Face** baseado na arquitetura **Transformers**. Basicamente, o modelo recebe uma mensagem incompleta com a sinalização de uma MASK (máscara) para que ele complete de acordo com as previsões já pré-treinadas do modelo.

Através da biblioteca `transformers`, o modulo `pipeline` será responsável por executar uma tarefa de PLN com uma alta abstração. Após isso, é só criar uma variável para carregar o modelo e anexa o texto. Assim é possível desenvolver previsões que preencham corretamente as frases.

```python
import transformers
from transformers import pipeline
```

É necessário ir a biblioteca da [Hugging Face](https://huggingface.co/) para encontrar um modelo que seja compatível com a tarefa que deseja executar. Nesse projeto, será usada a biblioteca `neuralmind/bert-large-portuguese-cased` para ter acesso a ela é só entrar no site e seguir os passos:

* *Hugging Face > Models > em Tasks vai em Natural Language Processing > seleciona Fill-Mask > em Languages seleciona Portuguese.*

Caso não coloque um modelo, será utilizado um modelo padrão que também funciona para a tarefa, porém a precisão (score) muda.

```python
mascarar = pipeline("fill-mask", model="neuralmind/bert-large-portuguese-cased")
```

```python
texto = mascarar("Talvez eu consiga [MASK] de casa mais cedo amanhã")
#texto = mascarar("Existe uma chance do copo cair no [MASK]")
#texto = mascarar("Bom [MASK], João! Como você está? Espero que você esteja bem hoje")
#texto = mascarar("Espero que João consiga um bom [MASK] para ter mais instabilidade na vida")

for x in range(len(texto)):
  print(texto[x])
```
