###Visão Geral
* Tarefa deste notebook: `summarization`
* Modelo utilizado: `facebook/bart-large-cnn`
* BART (Bidirectional and Auto-Regressive Transformers) é um modelo de linguagem desenvolvido pelo Facebook AI, que combina características de modelos auto-regressivos e bidirecionais, sendo eficaz para tarefas de geração de texto e compreensão de linguagem, como tradução automática e resumo de texto.

```python
#importando bibliotecas
import transformers
from transformers import pipeline
```

```python
resumo_pipeline = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
texto = """
A inteligência artificial (IA) refere-se à simulação da inteligência humana
em máquinas que são programadas para pensar como humanos e imitar suas ações.
O termo também pode ser aplicado a qualquer máquina que exiba traços associados
a uma mente humana, como aprendizagem e resolução de problemas.
 A IA é um campo interdisciplinar com várias abordagens, mas os avanços em
 aprendizado de máquina e aprendizado profundo estão criando uma mudança de
 paradigma em praticamente todos os setores da indústria de tecnologia.
 Desde a automação de tarefas repetitivas até a realização de diagnósticos
 médicos complexos, a IA está desempenhando um papel cada vez mais
 importante em nossas vidas.
"""
```

```python
#passando parâmetros
resumo = resumo_pipeline(
    texto,
    max_length=60,
    min_length=25,
    do_sample=False
)
```

```python
print("Resumo:", resumo[0]['summary_text'])
```
